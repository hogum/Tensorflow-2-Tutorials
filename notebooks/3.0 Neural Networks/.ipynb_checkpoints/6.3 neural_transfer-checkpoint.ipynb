# [Neural Style Transfer](https://arxiv.org/abs/1508.06576)

- Neural Style Transfer involves imposing the artistic style of one image onto another.
- The image we start with is called the **content image**
- The image we end up with is the **style reference image**
- The transformed image is known as the **input image** [*Takes inputs from two sources*]

Neural Style Transfer defines **2 loss functions**:

a) The difference between the content of the two images

b) Difference in terms of style of the two images

We shall use the **[VGG19](https://github.com/fchollet/deep-learning-models/blob/master/vgg19.py)** architecture for implementation

#### preprocess the images

```python
content_path = '.media/snap_s6.png' # content image
style_path = '.media/snap_spider.png' # style_image
```

```python
def load_image(image_path):
    """
        Reads an image file to an array
    """
    max_dims = 512
    img = PIL.Image.open(image_path)
    
    # resize
    longest_side = max(img.size)
    scale = max_dims / longest_side
    img = img.resize(size=(round(img.size[0] * scale),
                           round(img.size[1] * scale)),
                    resample=PIL.Image.ANTIALIAS)
    
    # cast to array
    img = tf.keras.preprocessing.image.img_to_array(img)
    img = np.expand_dims(img, axis=0)
    
    # Covert RGB channels to BGR
    img = tf.keras.applications.vgg19.preprocess_input(img)
    
    return img
```

```python
# Let's add a function to help us visualize the images
channel_means = [103.939, 116.779, 123.68] # BGR Channels means for VGR processing

def display_image(image, title=None, revert_channels=False):
    # remove the batch dimension
    if len(image.shape) == 4:
        image = np.squeeze(image, axis=0).astype('uint8')
    image = image.astype('float64')
    
    # Convert BGR to RGB
    image[:, :, 0] += channel_means[0] # these are the means subtracted by the preprocessing step
    image[:, :, 1] += channel_means[1]
    image[:, :, 2] += channel_means[2]
    
    image = image[:, :, ::-1].astype('uint8')
    
    if revert_channels:
        return
    plt.imshow(image)
    if title:
        plt.title(title)
    
```

#### Define the model

**Content layers**

The highest level is used because higher levels capture high level content [objects and their arrangement on the input image and  place no constrain on the actual exact pixels of the reconstruction

****

```python
content_layers = ['block5_conv2']

style_layers = ['block1_conv1',
               'block2_conv1',
               'block3_conv1',
               'block4_conv1',
               'block5_conv1']
```

```python
def get_model():
    """
        Returns the model that outputs the style and content
        and intermediate layers of the trained VGG19 model 
    """
    model = tf.keras.applications.vgg19.VGG19(include_top=False, weights='imagenet')
    model.trainable=False
    
    # Acquire outputs corresponding to layers
    style_outputs = [model.get_layer(name).output for name in style_layers]
    content_outputs = [model.get_layers(name).output for name in content_layers]
    
    outputs = style_outputs + content_outputs
    
    return tf.keras.models.Model(model.input, outputs)
```

#### Find Losses

```python
# losses

def find_rms_loss(img1, img2):
    """
        Losses between the content and styles of the two images
    """
    return tf.reduce_mean(tf.square(img1 - img1))

def content_loss(content, target):
    """
        Mean squared difference between loss and target
    """
    return find_rms_loss(content, target)
```

- A **gram matrix** is a dot product of the style matrix with its transpose. It defines the style loss
- It distributes the spatial information in an image to give its style: texture, weights and shapes

```python
def get_gram_matrix(input_tensor):
    
    # reshape to [channels, width, height]
    img_channels = int(inpu.shape[-1])
    tensor = tf.reshape(input_tensor, [-1, img_channels])
    
    n_channels = tf.shape(input=tensor)[0]
    
    gram = tf.matmul(tensor, tensor, transpose_a=True) # tensor^t * tensor
    
    return gram / n_channels


def style_loss(style, gram_target):
    gram_style = get_gram_matrix(style)
    return find_rms_loss(gram_style, gram_target)
    
```

By feeding the content image and the style image through model, we find the **content features** and **style features**

```python
def get_feature_representations(model):
    # Contents: Call the model on the image and extract feature layers
    
    content_image = load_image(content_path)
    content_ouputs = get_model(content_image)
    cont_features = [layer[0] for layer in content_ouputs[len(content_layers):]]
    
    style_image = load_image(style_path)
    style_outputs = get_model(style_image)
    style_features = [layer[0] for layer in style_outputs[:len(style_layers)]]
    
    return style_features, cont_features
```

```python
def find_loss(model, loss_weights, input_img, gram_style_features, content_features):
    """
        Finds the content, style, and total loss
    """
    content_score, style_score = 0, 0
    n_content = len(content_layers)  # Number of layers
    n_style =  len(style_layers)
    
    style_weight, content_weight = loss_weights
    
    # Call the model on the initial image
    outputs = model(input_img)
    
    
    # Get the content representations in desired layers
    content_output_features = outputs[-n_content:]
    weights_per_content_layer = 1 / n_content
    
    # Accumulate the content score for each layer
    for target, combination in zip(content_features, content_output_features):
        c_loss = content_loss(combination[0], target)
        content_score += weights_per_content_layer * c_loss
    
    # Weight the loss contribution of each layer
    content_score *= content_weight
    
    # Get style representations and accumulate weighted layer losses
    
    style_outputs = outputs[:n_style]
    
    w_style_lr = 1 / n_style # weights_per_style_layer 
    
    for target, combination in zip(gram_style_features, style_outputs):
        style_score += w_style_lr * style_loss(combination[0], target)
    style_score *= w_style_lr
    
    loss = style_score + content_score
    
    return loss, content_score, style_score
        
    
    
```

```python
def grad(**kwargs):
    """
        Computes gradients w.r.t the input image
    """
    with tf.GradientTape() as tape:
        loss = find_loss(**kwargs)
    total_loss = loss[0]
    
    return tape.gradient(total_loss, kwargs.get('input_image')), loss
        
```

#### Perform the Style Transfer

```python
def transfer_style(content_weight=1e3, style_weight=1e-2, n_iters=1000, displ_interval=10):
    lowest_loss = float('inf')
    best_image = None
    input_images = []
    
    model = get_model()
    
    # Only using output values from the layers: no need to train
    for layer in model.layers:
        layer.trainable = False
    
    style_features, content_features = get_feature_representations(model)
    
    # Add gram style features
    gram_style_features = [get_gram_matrix(feature) for feature in style_features]
    
    # Initialize the intented final image(Input image)
    content_img = load_image(content_path)
    content_img = tf.Variable(content_img, dtype=tf.float32)
        
    loss_w = style_weight, content_weight
    
    # calculate image bounds
    norm_means = np.asarray(channel_means)
    min_vals = -norm_means
    max_vals = 255 - norm_means
    
    config = {
        'model': model,
        'loss_weights': loss_w,
        'input_image': content_img,
        'gram_style_features': gram_style_features,
        'content_features': content_features
    }
    
    for i in range(n_iters):
        losses, img_ = apply_grads(config, min_vals, max_vals)
        t_loss, c_loss, s_loss = losesses
        
        if t_loss < lowest_loss:
            lowest_loss  = t_loss
            best_image = img_.numpy()
            
        print(f'{i}   loss: {t_loss:.5e} content loss:  {c_loss:.5e}  style_loss: {s_loss:.5e}')

        # Save images for display
        if not i % displ_interval:
            save_image(input_images, img_.numpy())
    
    # display best images
    display_ouput(input_images)

    return best_image,lowest_loss
            
```
